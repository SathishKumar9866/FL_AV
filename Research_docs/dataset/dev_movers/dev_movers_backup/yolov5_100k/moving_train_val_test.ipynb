#### dividing the dataset into three equal parts using random selection.

```python
# Source folder path where the images are located
source_folder_train = folder_path + os.sep + "train"
print(source_folder_train)
source_folder_val = folder_path + os.sep + "val"
# Destination folder path where the selected images will be copied
destination_folder = r"C:\Users\SIU856522160\Downloads\major\data"
```

#### number named annotations in given label in bddk100 website are 69863. we need to equally partition the data into dataset.

```python
# Read file names from the train.txt file
with open("output.txt", "r") as train_file:
    file_list = train_file.read().splitlines()

# Determine the total number of files and the desired number of groups
num_files = len(file_list)
m = 100

# Randomly divide the total number of files into groups
remaining_files_val = num_files % m
print(remaining_files_val,num_files)
file_groups = [random.sample(file_list, num_files // m) for _ in range(m)]
file_groups[-1].extend(random.sample(file_list, remaining_files_val))

# Get the list of device folders
device_folders = [folder for folder in os.listdir(destination_folder) if os.path.isdir(os.path.join(destination_folder, folder))]

# Assign each group to the same device sequentially
for group_index, file_group in enumerate(file_groups):
    device_index = group_index % len(device_folders)
    device_folder = device_folders[device_index]
    device_destination_folder = os.path.join(destination_folder, device_folder)

    # Create the destination folder for the current device if it doesn't exist
    os.makedirs(device_destination_folder, exist_ok=True)
    # Create the destination folder for the current device if it doesn't exist
    os.makedirs(device_destination_folder, exist_ok=True)

    # Delete existing files in train and val sub folders
    train_images_folder = os.path.join(device_destination_folder, "images", "train")
    train_labels_folder = os.path.join(device_destination_folder, "labels", "train")
    val_images_folder = os.path.join(device_destination_folder, "images", "val")
    val_labels_folder = os.path.join(device_destination_folder, "labels", "val")

    if os.path.exists(train_images_folder):
        with os.scandir(train_images_folder) as entries:
            for entry in entries:
                if entry.is_file():
                    os.remove(entry.path)
        print("Deleted files in train images folder")

    if os.path.exists(train_labels_folder):
        with os.scandir(train_labels_folder) as entries:
            for entry in entries:
                if entry.is_file():
                    os.remove(entry.path)
        print("Deleted files in train labels folder")

    if os.path.exists(val_images_folder):
        with os.scandir(val_images_folder) as entries:
            for entry in entries:
                if entry.is_file():
                    os.remove(entry.path)
        print("Deleted files in val images folder")

    if os.path.exists(val_labels_folder):
        with os.scandir(val_labels_folder) as entries:
            for entry in entries:
                if entry.is_file():
                    os.remove(entry.path)
        print("Deleted files in val labels folder")

    # Create train and val subfolders for images and labels
    os.makedirs(train_images_folder, exist_ok=True)
    os.makedirs(train_labels_folder, exist_ok=True)
    os.makedirs(val_images_folder, exist_ok=True)
    os.makedirs(val_labels_folder, exist_ok=True)
    
    # Create a text file to store the train file names for the current group
    train_group_txt_path = os.path.join(device_destination_folder, f"group_{group_index}_train.txt")
    label_group_txt_path = os.path.join(device_destination_folder, f"group_{group_index}_labels.txt")

    with open(train_group_txt_path, "w") as train_group_txt_file, open(label_group_txt_path, "w") as label_group_txt_file:
        for file_name in file_group:
            train_group_txt_file.write(f"{file_name}\n")
            label_name = file_name.split(".")[0] + ".txt"
            label_group_txt_file.write(f"{label_name}\n")

    print(f"Group {group_index}: Train files count: {len(file_group)}, Device: {device_folder}")
```

```python
group_index = 0
# Get the list of device folders
device_folders = [folder for folder in os.listdir(destination_folder) if os.path.isdir(os.path.join(destination_folder, folder))]
# print(device_folder)
# print(destination_folder)
# Iterate over each device folder
for device_folder in device_folders:
    device_destination_folder = os.path.join(destination_folder, device_folder)
    # print(device_destination_folder)
    # print(type(device_folder))
    # Find the corresponding group train file for the current device
    group_train_file = os.path.join(device_destination_folder, f"group_{group_index}_train.txt")
    print(group_train_file)
    # Read the file names from the group train file
    with open(group_train_file, "r") as train_group_file:
        file_list = train_group_file.read().splitlines()
    # Create a file to store the names of the copied images
    file_name_img_save = f"group_saved_img_{group_index}_train.txt"
    file_name_txt_save = f"group_saved_txt_{group_index}_train.txt" 
    copied_images_file_img = open(f"{device_destination_folder}/{file_name_img_save}", "w")
    copied_images_file_txt = open(f"{device_destination_folder}/{file_name_txt_save}", "w")
    # Copy the images from the current group back to the source folder
    for file_name in file_list:
        image_source_path = os.path.join(source_folder_train, file_name)
        image_destination_path = os.path.join(device_destination_folder, "images", "train", file_name)

        try:
            shutil.copy2(src=image_source_path, dst=image_destination_path)
            print(f"Moved image: {file_name}")
            copied_images_file_img.write(f"{file_name}\n")
        except FileNotFoundError:
            print(f"Image not found: {file_name}")
            continue

        # Copy the corresponding label file if it exists
        label_name = file_name.split(".")[0] + ".txt"
        label_source_path = os.path.join(source_folder_train, label_name)
        label_destination_path = os.path.join(device_destination_folder, "labels", "train", label_name)

        if os.path.exists(label_source_path):
            try:
                shutil.copy2(src=label_source_path, dst=label_destination_path)
                print(f"Moved label: {label_name}")
                copied_images_file_txt.write(f"{file_name}\n")
            except FileNotFoundError:
                print(f"Label not found: {label_name}")
                continue
            
    # Close the file after writing
    copied_images_file_img.close()
    copied_images_file_txt.close()
    group_index+=1
```
