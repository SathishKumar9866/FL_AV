```python
def del_folders_and_files(destination_folder):

    if os.path.exists(destination_folder):
        for root, dirs, files in os.walk(destination_folder):
            for dir in dirs:
                subfolder_path = os.path.join(root, dir)
                try:
                    shutil.rmtree(subfolder_path)
                    print(f"Deleted subfolder: {subfolder_path}")
                except OSError as e:
                    print(f"Failed to delete subfolder: {subfolder_path} - {e}")
            for file in files:
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"Deleted file: {file_path}")
                except OSError as e:
                    print(f"Failed to delete file: {file_path} - {e}")
        print("Deleted sub folders and files in the destination folder")
    else:
        print("Destination folder does not exist")

```

```python
def create_batches(source_folder, num_batches, destination_folder, folder_name):
    # delete 
    del_folders_and_files(destination_folder)
    # Get a list of files in the source folder
    if "train" in folder_name.lower() or "val" in folder_name.lower():
        print(f"Folder is {folder_name}")
        folder_name = folder_name.lower()
    else:
        print("Folder is not train or val")
        return "given the correct folder name"
    
    # Replace with your folder name

    print(folder_name)
    file_path = os.path.join( f"output_{folder_name}.txt")

    with open(file_path, "r") as file:
        file_list = file.read().splitlines()
    total_files = len(file_list)
    print(total_files)
    files_per_batch = total_files // num_batches   # Limit num_files_per_batch to the total number of files
    print(files_per_batch)
    remaining_files = total_files % num_batches  # Calculate the remaining files
    print(remaining_files)
    # Create destination folders for each batch
    for batch_num in range(1, num_batches + 1):
        batch_folder = os.path.join(destination_folder, f"Batch_{batch_num}")
        os.makedirs(batch_folder)
        image_destination = os.path.join(batch_folder, "images", folder_name)
        label_destination = os.path.join(batch_folder, "labels", folder_name)
        if not os.path.exists(image_destination):
            os.makedirs(image_destination)
        if not os.path.exists(label_destination):
            os.makedirs(label_destination)
        
        batch_files = files_per_batch
        if remaining_files > 0:
            batch_files += 1
            remaining_files -= 1
        # Select random files for the current batch
        selected_files = random.sample(file_list, batch_files)
        print(len(selected_files))
        
        # Remove the selected files from the file list
        for file_name in selected_files:
            file_list.remove(file_name)

        # Create a file to store the names of the copied images
        file_name_img_save = f"batches_saved_img_{batch_num}_{folder_name}.txt"
        file_name_txt_save = f"batches_saved_txt_{batch_num}_{folder_name}.txt"
        copied_images_file_img = open(os.path.join(batch_folder, file_name_img_save), "w")
        copied_images_file_txt = open(os.path.join(batch_folder, file_name_txt_save), "w")

        # Copy the selected files to the respective batch folder
        for file_name in selected_files:
            image_source_path = os.path.join(source_folder, folder_name, file_name)
            print(image_source_path)
            image_destination_path = os.path.join(batch_folder, "images", folder_name, file_name)
            print(image_destination_path)
            try:
                shutil.copy2(src=image_source_path, dst=image_destination_path)
                print(f"Moved image: {file_name}")
                copied_images_file_img.write(f"{file_name}\n")
            except FileNotFoundError:
                print(f"Image not found: {file_name}")
                continue

            # Copy the corresponding label file if it exists
            label_name = file_name.split(".")[0] + ".txt"
            print(label_name)
            label_source_path = os.path.join(source_folder, folder_name, label_name)
            print(label_source_path)
            label_destination_path = os.path.join(batch_folder, "labels", folder_name, label_name)
            print(label_destination_path)
            if os.path.exists(label_source_path):
                try:
                    shutil.copy(src=label_source_path, dst=label_destination_path)
                    print(f"Moved label: {label_name}")
                    copied_images_file_txt.write(f"{label_name}\n")
                except FileNotFoundError:
                    print(f"Label not found: {label_name}")
                    continue
                except OSError as e:
                    print(f"Error occurred while copying label: {label_name} - {e}")
                    continue


        # Close the files after writing
        copied_images_file_img.close()
        copied_images_file_txt.close()

```

```python
batch_folder = destination_folder # Replace with the path to the batch folder
destination_folder_vech =  r"C:\Users\SIU856522160\Downloads\major\vech" # Replace with the path to the destination folder
del_folders_and_files(destination_folder_vech)
print(8*"*",end="\n")
print(8*"*",end="\n")
# Create three separate folders for each batch
for i in range(1, 4):
    folder_name = f"Device_{i}"
    folder_path = os.path.join(destination_folder_vech, folder_name)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path, exist_ok=True)
        print(f'{folder_path} created')
print(8*"*",end="\n")
print(8*"*",end="\n")
# Get a list of directories in the batch folder
directories = [entry.name for entry in os.scandir(batch_folder) if entry.is_dir()]
# Shuffle the directories randomly
random.shuffle(directories)

# Calculate the number of directories per destination folder
num_directories = len(directories)
directories_per_folder = num_directories // 3
# Randomly move the batch folders to three separate destination folders
for i, directory in enumerate(directories):
    source_path = os.path.join(batch_folder, directory)
    # print(source_path)
    print(8*"*",end="")
    device_number = (i // directories_per_folder) + 1
    destination_folder_name = f"Device_{device_number}"
    # print(destination_folder_name)
    
    destination_folder_path = os.path.join(destination_folder_vech, destination_folder_name)
    # print(destination_folder_path)
    if not os.path.exists(destination_folder_path):
        os.makedirs(destination_folder_path, exist_ok=True)
    destination_path = os.path.join(destination_folder_path, directory)
    
    shutil.move(source_path, destination_path)
    print(f"{directory} >>>> {destination_folder_name}")
print(8*"*",end="\n")
print(8*"*",end="\n")
print("Batch folders moved to separate destination folders successfully.")
```
